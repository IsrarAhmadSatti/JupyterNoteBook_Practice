```python
studentMarks = int(input("Enter your Marks"))

if (studentMarks <= 50):
    print("You FAILED!")
elif (50 < studentMarks <= 60):
    print("Your Grade is D")
elif (60 < studentMarks <= 70):
    print("Your Grade is C")
elif (70 < studentMarks <= 80):
    print("Your Grade is B")
elif (80 < studentMarks <= 90):
    print("Your Grade is A")
else:
    print("Your Grade is A+")
```

```python
class Car:
    def __init__ (self,color):
        self.color = color

    def start(self):
        print(f"{self.color} car started")

myCar = Car("red")
myCar.start()

secCar = Car("bro's")
secCar.start()
```

```python
class Car:
    color = "BLACK"
    def drive(self):
        print(f"{self.color} Car is moving")

    def setColor(self,newColor):
        self.color = newColor

car1 = Car()
car1.drive()

car2 = Car()
car2.setColor("GREEN")
car2.drive()
```

```python
class Student:
    def __init_(self, name, age, grade):
        self.name = name
        self.age = age
        self.grade = grade

    def display_info(self):
        print(f"Student Name: {self.name}")
        print(f"Student Age: {self.age}")
        print(f"Student Grade: {self.grade}")

    def is_eligible(self):
        if self.age >= 15:
            print(f"{self.name} is eligile for admission")
        else:
            print(f"{self.name} is NOT eligile for admission")
            print("Student Only above 15 year are eligble")

std1 = Student()
std1.name = str(input("Enter youe Name"))
std1.age = int(input("Enter your Age"))
std1.grade = str(input("Enter youe Grade"))
std1.is_eligible()
std1.display_info()
```
